```python
import numpy as np 
from sklearn.decomposition import PCA

X = np.load("data/.train_enc_hkpln.npy")

pos = np.load("data/.refs_positive.npy")
neg = np.load("data/.refs_negative.npy")
```

```python
pca = PCA(n_components=2)
tr = pca.fit_transform(X)

pos = pca.transform(pos)
neg = pca.transform(neg)
```

```python
import matplotlib.pyplot as plt

import pandas as pd

y = pd.read_csv("data/train_tmp.csv")["humor"].to_numpy()

%matplotlib.inline
```

```python
plt.scatter(tr[y == 0,0], tr[y == 0,1])
plt.scatter(tr[y == 1,0], tr[y == 1,1])

plt.scatter(pos[:,0], pos[:,1])
plt.scatter(neg[:,0], neg[:,1])
plt.show()
```
